```python
# creating the dataset
data = {'Ford':11, 'Tesla':19, 'BMW':5,
        'Mercedes':40,'JaGuar':25}
cars = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(cars, values, color ='maroon',
        width = 0.4)

plt.xlabel("CArs before 2010")
plt.ylabel("percentage of owners")
plt.title("Before 2010")
plt.show()
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(cars, values, color ='maroon')
# Add x, y gridlines
ax.grid(b = True, color ='black',
        linestyle ='-.', linewidth = 1,
        alpha = 0.2)
ax.invert_yaxis()
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='black')
######################################################################
# creating the dataset
data = {'Ford':38, 'Tesla':12, 'BMW':30,
        'Mercedes':14,'JaGuar':6}
cars = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(cars, values, color ='maroon',
        width = 0.4)

plt.xlabel("CArs After 2010")
plt.ylabel("percentage of owners")
plt.title("After 2010")
plt.show()
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(cars, values, color ='maroon')
# Add x, y gridlines
ax.grid(b = True, color ='black',
        linestyle ='-.', linewidth = 1,
        alpha = 0.2)
ax.invert_yaxis()
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='black')
```
